# Logistic Regression

Linear regression is the simplest supervised learning approach to train a continouos model.  
But what about the E-mail spam filter?

Email:	Spam Not Spam?	
Tumor:	Malignant /	Benign?

$$ y \in {0,1} $$

0:	“Negative	Class”	(e. g.,	benign	tumor)  
1:	“Positive	Class”	(e. g.,	malignant tumor)

$$ y \in {0, 1, 2, 3} $$

multiple diagnostic classes

Suppose our data is distributed like that:

```python
Image("regr8.png")
```

clearly, we want to distringuish between $y \leq 0.5$ and $y > 0.5$

the dashed orange line is a linear model while the green line is a *sigmoid function*:
$$ g(z) = \frac{1}{1+e^{-z}}$$

The *sigmoid* or *logistic* function allows to formulate our hypothesis $h_\theta(x)$ as:
$$h_\theta(x) = g(\theta^T x) $$
in a way that predicted y will be separated in two groups or *classes*

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(x):
    return 1.0 / (1 + np.exp(-x))

x = np.arange(-10,10,.1)
y = np.array(list(map(sigmoid, x)))
plt.plot(x, y)

plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sigmoid')
plt.show()
```

**Interpretation of the Hypothesis**:
$h_\theta$ is the estimation that y will be 1 (or 0) given some continuous input x and a model $\theta$:
$$  h_\theta(x) = P(y=1|x,\theta) $$

by the definition of probability:
$$ 
P(y=0)+P(y=1) =1 \\
P(y=0|x,\theta) = 1 - P(y=1|x,\theta) 
$$

# Logistic regression cost function

A sum of squared errors in this case would not give a convex shape, apt to be minimized by Gradient descent:  
we need another definition of J.

$$ J= \frac{1}{m} \sum_{i=1}^m Cost (h_\theta(x),y) $$

$$ Cost(h_\theta(x),y)= -log(h_\theta(x))$$  if $y=1$  
$$ Cost(h_\theta(x),y)= -log(1-h_\theta(x))$$  if $y=0$  


when $y=1$ we have $cost=0$ at $h_\theta=1$ and then we penalize with an increasing cost any prediction made with $h_\theta$ that goes beyond 0;  
on the contrary, when $y=0$ every value of $h_\theta$ that approaches $1$ gives an infinite cost.

More succinctly, we can write this as

$$ Cost(h_\theta(x),y)=−y log(h_\theta(x)) − (1−y) log(1−h_\theta(x)) $$

since the second term will be zero when y=1 and the first term will be zero when y=0. Substituting this cost into our overall cost function we obtain:

$$ J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} [y^{(i)}log(h_\theta(x^{(i)})) + (1 - y^{(i)})log(1 - h_\theta(x^{(i)}))] $$

## Gradient descent 
repeat until convergence {

>$$ \theta_j = \theta_j - \alpha \frac{\partial}{\partial \theta_j} J(\theta) $$

}

let $h_i = 1/(1+e^(-x^{i}\theta)$ and note that $h^{\prime}=h(1-h)$
$$\begin{align}
\frac{\partial}{\partial \theta_j} J(\theta_j) = &  -\frac{1}{m} \sum_{i=1}^{m}[ y^{(i)}log(h_i) + (1-y^{(i)})log(1-h_i)]\\
= & -\frac{1}{m} \sum_{i=1}^{m}[ y^{(i)}\frac{h_i^{\prime}}{h_i} + (1-y^{(i)})\frac{-1}{1-h_i}h_i^{\prime}] \\
= & -\frac{1}{m} \sum_{i=1}^{m}[ y^{(i)}\frac{h_i(1-h_i)}{h_i} +(y^{(i)}-1)\frac{1}{1-h_i}h_i(1-h_i)] \\
= & -\frac{1}{m} \sum_{i=1}^{m}[ y^{(i)}(1-h_i)+(y^{(i)}-1)h_i \\
= & -\frac{1}{m} \sum_{i=1}^{m}[ y^{(i)} - y^{(i)}h_i+y^{(i)}h_i-h_i]\\
= & \frac{1}{m} \sum_{i=1}^{m}[h_i- y^{(i)}]
\end{align}$$
i. e.  
$$ \frac{\partial}{\partial \theta_j} J(\theta) = ( \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)} ) $$

*Algorithm looks identical to linear regression!*

Alternatives to gradient descent (more powerful and costly):
- Coniugate gradients
- BGFS
- L-BGFS

# Hands on

```python
data = np.loadtxt('microchip_tests.csv', delimiter=',')

# Filter data based on last column value
passed = data[np.where(data[:,2] == 1)]
failed = data[np.where(data[:,2] == 0)]

plt.plot(passed[:, 0], passed[:, 1], 'go')
plt.plot(failed[:, 0], failed[:, 1], 'r+')

plt.grid(True)
plt.xlabel('Microchip Test 1')
plt.ylabel('Microchip Test 2')
plt.title('Microchip Tests')
plt.show()
```

## Hypothesis

$$ h_\theta(x) = g(\theta_0 * x_0 + \theta_1 * x_1 + \theta_2 * x_2 + theta_3 * x_1^2 + \theta_4*x_2^2) $$ 
- $ x_0 = 1 $ intercept
- $ x_1 $ - Microchip Test 1  
- $ x_2 $ - Microchip Test 2  

Sigmoid function:  
$$ g(z) = \frac{1}{1+e^{-z}} $$ 
$z = X^{m x n} \theta^{n x 1}$

```python
def map_features(data_x, mypow):
    tmp = list()
    for i, I in enumerate(mypow):
        tmp.append(data_x[:,i]**I)
    tmp = np.asarray(tmp).T
    data_x = np.append(data_x, tmp,axis=1)    
    return data_x
```

```python
def cost_function(theta, X, y):  
    m  = X.shape[0]
    hx = sigmoid(X*theta)
      
    J = (1.0 / m) * (((-1 * y.T).dot(np.log(hx))) - ((1 - y.T).dot(np.log(1 - hx)))) 
    return J[0,0]
```

```python
def gradient_descent(alpha, theta, X, y):
    m     = X.shape[0]
    hx    = sigmoid(X*theta)    
    grad  = (1.0 / m) * np.dot(X.T,(hx - y))
    theta = theta - alpha*grad
    return theta
```

```python
def Run(niter, alpha, data):
    data_x = data[:,:2]
    data_y = data[:,2]
    
    data_x = map_features(data_x,(2,2))
    intercept = np.ones((data_x.shape[0],1))
    data_x = np.append(intercept,data_x,axis=1)   
    X = np.matrix(data_x)
    y = np.matrix(np.expand_dims(data_y, axis=0).T)

    theta = np.random.rand(data_x.shape[1],1)
    theta = np.matrix(theta)
    
    JL = list()
    J = cost_function(theta, X, y)
    print("Initial cost J={0:f}".format(J))
    
    n = 0
    while J > 0.01 and n < niter:
        JL.append(J)
        theta = gradient_descent(alpha, theta, X, y)
        J = cost_function(theta, X, y)
        if n % 2000 == 0:
            print("Step {0:d}, J={1:15.9f}".format(n, JL[n]))
        if n > 0 and JL[-1] > JL[-2]:
            print("Error: J is increasing, J_cur-J_past={0:f}".format(JL[-1]-JL[-2]))
            break
        n += 1
        
    return theta, J, X, y
```

**Run**

```python
theta, J, X, y = Run(2e4, 1e-1, data)
print(theta, J)
```
